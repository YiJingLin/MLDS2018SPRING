```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
batch = [10, 50, 100, 500, 1000, 6000]
Froberius_norm = [0.9122, 0.0979, 0.0442, 0.0514, 0.0241, 0.0094]
train_loss = [0.1146, 0.0539, 0.0346, 0.1101, 0.0418, 0.0562]
train_acc = [0.9687, 0.9602, 0.9761, 0.9789, 0.9876, 0.9836]
test_loss = [0.1480, 0.1804, 0.0883, 0.0766, 0.0567, 0.0619]
test_acc = [0.9490, 0.9460, 0.9700, 0.9750, 0.9815, 0.9790]
```

```python
fig, ax1 = plt.subplots()


# first dataset
ax1.plot(batch, train_acc,'b-')
ax1.plot(batch, test_acc,'b--')


ax1.set_xlabel('batch size')
ax1.set_xscale("log", nonposx='clip')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('accuracy', color='b')
ax1.tick_params('y', colors='b') # axis param

# second dataset
ax2 = ax1.twinx()
ax2.plot(batch, Froberius_norm, 'r-')
ax2.set_ylabel('sensitivity', color='r')
ax2.tick_params('y', colors='r') # axis param

fig.tight_layout()
plt.show()
fig.savefig('./acc-sensitivity.png')
```

```python
fig, ax1 = plt.subplots()


# first dataset
ax1.plot(batch, train_loss,'b-')
ax1.plot(batch, test_loss,'b--')


ax1.set_xlabel('batch size')
ax1.set_xscale("log", nonposx='clip')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('cross entropy loss', color='b')
ax1.tick_params('y', colors='b') # axis param

# second dataset
ax2 = ax1.twinx()
ax2.plot(batch, Froberius_norm, 'r-')
ax2.set_ylabel('sensitivity', color='r')
ax2.tick_params('y', colors='r') # axis param

fig.tight_layout()
plt.show()
fig.savefig('./loss-sensitivity.png')
```
