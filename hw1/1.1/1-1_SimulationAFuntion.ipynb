# 1-1 Simulate a funtion :
- function need to be single-input, single-output
- function need to be non-linear
- ex: y = tan(x)



```python
from keras.models import Model
from keras.layers import Input, Dense

import numpy as np
```

## create training data

```python
### create input data : X

interval = 0.001

X = np.arange(0., 1.+interval, interval)
X[:10]
```

```python
### create ouput data : Y

Y = np.sin(10 * np.pi * X) / 10 * np.pi * X
Y[:10]
```

## create model

```python
from keras import losses
from keras import optimizers

mse = losses.mean_squared_error
adam = optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0)


def create_shallow_model():
    input_layer = Input(shape=(1,))
    hidden_layer=Dense(300)(input_layer)
    output_layer = Dense(1)(hidden_layer)
    shallow_model = Model(inputs=input_layer, outputs=output_layer)

    shallow_model.summary()
    return shallow_model

def create_deep_model():
    input_layer = Input(shape=(1,))
    hidden_layer=Dense(37)(input_layer)
    for ite in range(4):
        hidden_layer=Dense(11)(hidden_layer)
    output_layer = Dense(1)(hidden_layer)
    model = Model(inputs=input_layer, outputs=output_layer)

    model.summary()
    return model
```

```python
### shallow model 

shallow_model = create_shallow_model()
shallow_model.compile(loss=mse, optimizer='adam', metrics=['accuracy'])
shallow_result = shallow_model.fit(x=X, y=Y, batch_size=64, epochs=1000)

shallow_prediction = shallow_model.predict(x=X)
```

```python
deep_model = create_deep_model()
deep_model.compile(loss=mse, optimizer='adam', metrics=['accuracy'])
deep_result = deep_model.fit(x=X, y=Y, batch_size=64, epochs=1000)

deep_prediction = deep_model.predict(x=X)
```

```python
### draw result
import matplotlib.pyplot as plt

plt.figure()

plt.title('Result')
plt.xlabel('X')
plt.ylabel('Y')

# ax = plt.gca().set_color_cycle(['red', 'green', 'blue'])

plt.plot(X, Y, label=r"\mathrm{sin}(2 \omega t)$")
plt.plot(X, deep_prediction)
plt.plot(X, shallow_prediction)


plt.legend([r'$\frac{\sin ({5}{\pi}{x})}{5 \pi x}$', 'deep', 'shallow'], loc='upper left')
plt.savefig('result.png')
plt.show()
```
